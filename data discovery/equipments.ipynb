```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
file_path = "Equipments.xlsx"
df = pd.read_excel(file_path)

# Colonnes numériques
numeric_cols = ["CO2_Emissions_kg", "Energy_Consumption_kWh", "Estimated_Lifetime_Years"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Détection des valeurs aberrantes
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
outlier_counts = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))).sum()

# Fonction d'annotation
def annotate_boxplot(ax, has_outliers):
    text = "Valeurs aberrantes détectées" if has_outliers else "Pas de valeurs aberrantes"
    ax.text(1.05, 0.95, text, transform=ax.transAxes, color="red" if has_outliers else "green", fontsize=10, verticalalignment='top')

# Création des boxplots
fig, axes = plt.subplots(1, len(numeric_cols), figsize=(15, 6))
for i, col in enumerate(numeric_cols):
    sns.boxplot(y=df[col], ax=axes[i], color='lightblue')
    axes[i].set_title(f"Boxplot de {col}")
    annotate_boxplot(axes[i], outlier_counts[col] > 0)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
file_path = "Equipments.xlsx"
df = pd.read_excel(file_path)

# Colonnes numériques
numeric_cols = ["CO2_Emissions_kg", "Energy_Consumption_kWh", "Estimated_Lifetime_Years"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Calcul de la matrice de corrélation pour les colonnes numériques
corr_matrix = df[numeric_cols].corr()

# Affichage de la matrice de corrélation avec heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matrice de corrélation des variables numériques")
plt.show()

# Interprétation des relations entre colonnes
print("Interprétation des relations entre colonnes :\n")
for i, col1 in enumerate(numeric_cols):
    for j in range(i + 1, len(numeric_cols)):
        col2 = numeric_cols[j]
        corr_value = corr_matrix.loc[col1, col2]
        # Définir une interprétation qualitative
        if abs(corr_value) >= 0.7:
            relation = "forte"
        elif abs(corr_value) >= 0.3:
            relation = "modérée"
        else:
            relation = "faible"
        # Indiquer le signe de la relation
        signe = "positive" if corr_value > 0 else "négative" if corr_value < 0 else "aucune"
        print(f"La corrélation entre '{col1}' et '{col2}' est de {corr_value:.2f} ({relation} {signe}).")
```

```python
import pandas as pd

# Charger les données
file_path = "Equipments.xlsx"
df = pd.read_excel(file_path)

# Création d'un Data Dictionary
data_dict = {
    "Nom de la Colonne": [],
    "Type de données": [],
    "Nombre de valeurs manquantes": [],
    "Exemples de valeurs": [],
    "Description": []
}

# Remplissage pour chaque colonne du DataFrame
for col in df.columns:
    data_dict["Nom de la Colonne"].append(col)
    data_dict["Type de données"].append(str(df[col].dtype))
    data_dict["Nombre de valeurs manquantes"].append(df[col].isnull().sum())
    exemples = df[col].dropna().astype(str).head(3).tolist()
    data_dict["Exemples de valeurs"].append(", ".join(exemples))

    # Description pour les colonnes présentes dans Equipments.xlsx
    if col == "Equipment_ID":
        desc = "Identifiant unique de l'équipement"
    elif col == "Equipment_Name":
        desc = "Nom de l'équipement"
    elif col == "Category":
        desc = "Catégorie de l'équipement"
    elif col == "Maintenance_Cycle":
        desc = "Fréquence de maintenance"
    elif col == "Location":
        desc = "Emplacement de l'équipement"
    elif col == "Manufacturer":
        desc = "Fabricant de l'équipement"
    elif col == "CO2_Emissions_kg":
        desc = "Émissions de CO2 (en kg)"
    elif col == "Energy_Consumption_kWh":
        desc = "Consommation d'énergie (en kWh)"
    elif col == "Energy_Type":
        desc = "Type d'énergie utilisée"
    elif col == "Maintenance_Frequency":
        desc = "Fréquence de maintenance"
    elif col == "Estimated_Lifetime_Years":
        desc = "Durée de vie estimée (en années)"
    else:
        desc = ""
    data_dict["Description"].append(desc)

# Conversion du dictionnaire en DataFrame
df_dict = pd.DataFrame(data_dict)

# Affichage du Data Dictionary
print(df_dict)

# Export vers Excel (le fichier source n'est pas modifié)
output_excel = "Data_Dictionary.xlsx"
df_dict.to_excel(output_excel, index=False)
print(f"\nData Dictionary exporté dans le fichier : {output_excel}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la page à scraper (vous pouvez changer cette URL pour d'autres pages de produits)
url = 'https://geneq.com/biotechnology/fr/alimentaire-equipements-167/'

# Faire une requête HTTP pour obtenir le contenu de la page
response = requests.get(url)
html_content = response.text

# Utiliser BeautifulSoup pour parser le HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Exemple : récupérer les titres des équipements, les prix et les liens des produits

# Trouver toutes les sections contenant des informations sur les équipements
items = soup.find_all('div', class_='listing-item')

# Liste pour stocker les données extraites
data = []

# Parcourir les éléments pour extraire les informations
for item in items:
    # Récupérer le nom du produit (titre)
    title = item.find('a', class_='listing-title')
    if title:
        title = title.text.strip()

    # Récupérer le prix
    price = item.find('span', class_='listing-price')
    if price:
        price = price.text.strip()

    # Récupérer l'URL du produit (lien)
    link = item.find('a', class_='listing-title')['href']
    if link:
        link = 'https://www.farmmachinerysales.com.au' + link

    # Ajouter les informations collectées à la liste des données
    data.append([title, price, link])

# Convertir la liste de données en DataFrame pour une manipulation plus facile
df = pd.DataFrame(data, columns=['Nom de l\'équipement', 'Prix', 'Lien'])

# Afficher les 10 premières lignes du DataFrame
print(df.head(10))

# Sauvegarder les résultats dans un fichier CSV
df.to_csv('equipements_agricoles_farmmachinerysales.csv', index=False)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données depuis un fichier Excel
data = pd.read_excel('Equipments.xlsx')

# 1. Aperçu des données
print("Aperçu des données:")
print(data.head())

# 2. Informations sur les colonnes et types de données
print("\nInformations sur les données:")
print(data.info())

# 3. Statistiques descriptives des colonnes numériques
print("\nStatistiques descriptives:")
print(data.describe())

# 4. Nombre de valeurs manquantes par colonne
print("\nNombre de valeurs manquantes par colonne:")
print(data.isnull().sum())

# 5. Distribution des valeurs manquantes
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title("Carte thermique des valeurs manquantes")
plt.show()

# 6. Distribution des variables numériques (si elles existent sous forme numérique)
# Convertir les colonnes numériques qui sont actuellement des chaînes de caractères
data['Unit_Price'] = pd.to_numeric(data['Unit_Price'], errors='coerce')
data['Stock_Initial'] = pd.to_numeric(data['Stock_Initial'], errors='coerce')
data['Total_Stock_Value'] = pd.to_numeric(data['Total_Stock_Value'], errors='coerce')
data['Carbon_Footprint_per_Unit_kgCO2e'] = pd.to_numeric(data['Carbon_Footprint_per_Unit_kgCO2e'], errors='coerce')
data['Transport_Distance_km'] = pd.to_numeric(data['Transport_Distance_km'], errors='coerce')
data['Water_Consumption_per_Unit_liters'] = pd.to_numeric(data['Water_Consumption_per_Unit_liters'], errors='coerce')

# Histogrammes pour la distribution des variables numériques
numerical_columns = ['Unit_Price', 'Stock_Initial', 'Total_Stock_Value', 'Carbon_Footprint_per_Unit_kgCO2e',
                     'Transport_Distance_km', 'Water_Consumption_per_Unit_liters']

data[numerical_columns].hist(bins=30, figsize=(12, 10))
plt.suptitle("Distribution des variables numériques")
plt.show()

# 7. Boxplots pour visualiser les valeurs aberrantes
plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot de {col}')
plt.tight_layout()
plt.show()

# 8. Analyse des relations entre certaines variables numériques
plt.figure(figsize=(10, 6))
sns.pairplot(data[numerical_columns])
plt.suptitle("Relation entre les variables numériques", y=1.02)
plt.show()

# 9. Comptage des catégories pour les variables catégorielles (exemple : Type, Recycled_Packaging)
categorical_columns = ['Type', 'Recycled_Packaging']
plt.figure(figsize=(10, 6))
for i, col in enumerate(categorical_columns, 1):
    plt.subplot(1, 2, i)
    sns.countplot(x=data[col])
    plt.title(f'Répartition des valeurs dans {col}')
plt.tight_layout()
plt.show()

# 10. Matrice de corrélation entre les variables numériques
plt.figure(figsize=(10, 6))
correlation_matrix = data[numerical_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Matrice de Corrélation")
plt.show()
```
