```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Charger les données
df = pd.read_json('data.json')

pd.set_option('future.no_silent_downcasting', False)

# Convertir la colonne en numérique (les valeurs non convertibles deviendront NaN)
df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')

# Liste des colonnes numériques
numeric_columns = ['Wind_Speed', 'Humidity', 'CO2_Ambient_Level', 'Temperature']

# Affichage des statistiques descriptives
stats = df[numeric_columns].describe()

# Tracer les boxplots en grille
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Boxplots des Variables Numériques", fontsize=16)

for ax, col in zip(axes.flatten(), numeric_columns):
    sns.boxplot(y=df[col], ax=ax)
    ax.set_title(f"{col}")
    ax.set_xlabel("")
    ax.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Interprétation des statistiques
for col in numeric_columns:
    print(f"--- {col} ---")
    print(f"Min : {stats.loc['min', col]}")
    print(f"Q1 (25%) : {stats.loc['25%', col]}")
    print(f"Médiane (50%) : {stats.loc['50%', col]}")
    print(f"Q3 (75%) : {stats.loc['75%', col]}")
    print(f"Max : {stats.loc['max', col]}\n")

    # Calcul de l'IQR (Interquartile Range)
    IQR = stats.loc['75%', col] - stats.loc['25%', col]
    print(f"IQR (Intervalle Interquartile) : {IQR}")

    # Détection des valeurs aberrantes
    lower_bound = stats.loc['25%', col] - 1.5 * IQR
    upper_bound = stats.loc['75%', col] + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"Présence de valeurs aberrantes : {'Oui' if not outliers.empty else 'Non'}")
    print(f"Nombre de valeurs aberrantes : {outliers.shape[0]}")
    print("\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Charger les données
df = pd.read_json('data.json')

# Liste des colonnes numériques
numeric_columns = ['Temperature', 'Humidity', 'CO2_Ambient_Level', 'Wind_Speed']

# Convertir les colonnes numériques en valeurs numériques (en remplaçant les erreurs par NaN)
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Gérer les valeurs manquantes (ici on peut choisir de les remplir par la moyenne ou la médiane)
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())  # Remplir par la moyenne

# Calcul de la matrice de corrélation
corr_matrix = df[numeric_columns].corr()

# Affichage sous forme de heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Matrice de Corrélation")
plt.show()

# Interprétation des corrélations
def interpret_correlation(value):
    """Retourne une interprétation de la corrélation en fonction de sa valeur."""
    if value > 0.7:
        return "Forte corrélation positive : les deux variables évoluent ensemble."
    elif value > 0.4:
        return "Corrélation modérée : il y a une tendance mais pas systématique."
    elif value > 0.1:
        return "Corrélation faible : peu d'influence entre ces variables."
    elif value < -0.7:
        return "Forte corrélation négative : quand l'une augmente, l'autre diminue fortement."
    elif value < -0.4:
        return "Corrélation modérée négative : tendance inverse présente."
    elif value < -0.1:
        return "Corrélation faible négative : influence limitée."
    else:
        return "Aucune corrélation significative."

# Affichage de l'interprétation
for col1 in numeric_columns:
    for col2 in numeric_columns:
        if col1 != col2:
            corr_value = corr_matrix.loc[col1, col2]
            print(f"Corrélation entre {col1} et {col2} : {corr_value:.2f} → {interpret_correlation(corr_value)}\n")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les données
df = pd.read_json('data.json')



# Conversion de la colonne Date en format datetime
df['Date'] = pd.to_datetime(df['Date'])

# Nettoyage des valeurs non numériques dans Temperature
df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')

# Suppression des lignes avec des valeurs NaN après conversion
df = df.dropna(subset=['Temperature'])



# Style des graphiques
sns.set_theme(style="darkgrid")

# Création des figures
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Distribution des températures
sns.histplot(df["Temperature"], bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Distribution des températures")

# 2. Évolution temporelle de la température
sns.lineplot(data=df, x="Date", y="Temperature", ax=axes[0, 1])
axes[0, 1].set_title("Évolution temporelle de la température")
axes[0, 1].tick_params(axis='x', rotation=45)

# 3. Matrice de corrélation
corr_matrix = df.select_dtypes(include=['float64']).corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", ax=axes[1, 0])
axes[1, 0].set_title("Matrice de corrélation")

# 4. Distribution des niveaux de CO₂
sns.histplot(df["CO2_Ambient_Level"], bins=30, kde=True, ax=axes[1, 1])
axes[1, 1].set_title("Distribution des niveaux de CO2")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Charger les données
df = pd.read_json('../data.json')

# Initialiser un dictionnaire pour stocker les valeurs mal formatées
improper_formatting = {}

# 1. Vérification des colonnes numériques
for col in ['Temperature', 'Humidity', 'Wind_Speed', 'CO2_Ambient_Level']:
    # Conversion en float avec erreurs converties en NaN
    df[col + '_converted'] = pd.to_numeric(df[col], errors='coerce')
    # Filtrer les valeurs mal formatées (NaN après conversion mais non NaN initialement)
    improper_formatting[col] = df[df[col + '_converted'].isna() & df[col].notna()][col]

# 2. Vérification de la colonne Date
df['Date_converted'] = pd.to_datetime(df['Date'], errors='coerce')
improper_formatting['Date'] = df[df['Date_converted'].isna() & df['Date'].notna()]['Date']

# 3. Vérification des colonnes catégoriques (valeurs non string ou nulles)
improper_formatting['Location'] = df[~df['Location'].apply(lambda x: isinstance(x, str) or pd.isna(x))]['Location']

# 4. Affichage des résultats
for col, values in improper_formatting.items():
    print(f"\nValeurs mal formatées dans la colonne '{col}' :\n{values if not values.empty else 'Aucune valeur mal formatée trouvée'}")
```

```python
import pandas as pd

# Charger les données
df = pd.read_csv('../Materials.txt')

# Fonction pour vérifier les valeurs mal formatées dans chaque colonne
def check_improper_formatting(df):
    improper_formatting_counts = {}

    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            # Vérifier les valeurs non numériques dans une colonne supposée être numérique
            non_numeric_values = df[~df[column].apply(pd.to_numeric, errors='coerce').notna()]
            improper_formatting_counts[column] = non_numeric_values.shape[0]
        else:
            # Vérifier les espaces en début ou fin pour les colonnes de type chaîne de caractères
            improper_formatting_counts[column] = df[column].apply(lambda x: isinstance(x, str) and (x.strip() != x) if pd.notnull(x) else False).sum()

    # Afficher les résultats
    print("\n--- Nombre de valeurs mal formatées par colonne ---")
    for column, count in improper_formatting_counts.items():
        print(f"{column}: {count} valeur(s) mal formatée(s)")

# Appeler la fonction pour générer le rapport sur les valeurs mal formatées
check_improper_formatting(df)
```
