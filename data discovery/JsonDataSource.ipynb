```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Charger les donnÃ©es
df = pd.read_json('data.json')

pd.set_option('future.no_silent_downcasting', False)

# Convertir la colonne en numÃ©rique (les valeurs non convertibles deviendront NaN)
df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')

# Liste des colonnes numÃ©riques
numeric_columns = ['Wind_Speed', 'Humidity', 'CO2_Ambient_Level', 'Temperature']

# Affichage des statistiques descriptives
stats = df[numeric_columns].describe()

# Tracer les boxplots en grille
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Boxplots des Variables NumÃ©riques", fontsize=16)

for ax, col in zip(axes.flatten(), numeric_columns):
    sns.boxplot(y=df[col], ax=ax)
    ax.set_title(f"{col}")
    ax.set_xlabel("")
    ax.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# InterprÃ©tation des statistiques
for col in numeric_columns:
    print(f"--- {col} ---")
    print(f"Min : {stats.loc['min', col]}")
    print(f"Q1 (25%) : {stats.loc['25%', col]}")
    print(f"MÃ©diane (50%) : {stats.loc['50%', col]}")
    print(f"Q3 (75%) : {stats.loc['75%', col]}")
    print(f"Max : {stats.loc['max', col]}\n")

    # Calcul de l'IQR (Interquartile Range)
    IQR = stats.loc['75%', col] - stats.loc['25%', col]
    print(f"IQR (Intervalle Interquartile) : {IQR}")

    # DÃ©tection des valeurs aberrantes
    lower_bound = stats.loc['25%', col] - 1.5 * IQR
    upper_bound = stats.loc['75%', col] + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"PrÃ©sence de valeurs aberrantes : {'Oui' if not outliers.empty else 'Non'}")
    print(f"Nombre de valeurs aberrantes : {outliers.shape[0]}")
    print("\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Charger les donnÃ©es
df = pd.read_json('data.json')

# Liste des colonnes numÃ©riques
numeric_columns = ['Temperature', 'Humidity', 'CO2_Ambient_Level', 'Wind_Speed']

# Convertir les colonnes numÃ©riques en valeurs numÃ©riques (en remplaÃ§ant les erreurs par NaN)
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# GÃ©rer les valeurs manquantes (ici on peut choisir de les remplir par la moyenne ou la mÃ©diane)
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())  # Remplir par la moyenne

# Calcul de la matrice de corrÃ©lation
corr_matrix = df[numeric_columns].corr()

# Affichage sous forme de heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Matrice de CorrÃ©lation")
plt.show()

# InterprÃ©tation des corrÃ©lations
def interpret_correlation(value):
    """Retourne une interprÃ©tation de la corrÃ©lation en fonction de sa valeur."""
    if value > 0.7:
        return "Forte corrÃ©lation positive : les deux variables Ã©voluent ensemble."
    elif value > 0.4:
        return "CorrÃ©lation modÃ©rÃ©e : il y a une tendance mais pas systÃ©matique."
    elif value > 0.1:
        return "CorrÃ©lation faible : peu d'influence entre ces variables."
    elif value < -0.7:
        return "Forte corrÃ©lation nÃ©gative : quand l'une augmente, l'autre diminue fortement."
    elif value < -0.4:
        return "CorrÃ©lation modÃ©rÃ©e nÃ©gative : tendance inverse prÃ©sente."
    elif value < -0.1:
        return "CorrÃ©lation faible nÃ©gative : influence limitÃ©e."
    else:
        return "Aucune corrÃ©lation significative."

# Affichage de l'interprÃ©tation
for col1 in numeric_columns:
    for col2 in numeric_columns:
        if col1 != col2:
            corr_value = corr_matrix.loc[col1, col2]
            print(f"CorrÃ©lation entre {col1} et {col2} : {corr_value:.2f} â†’ {interpret_correlation(corr_value)}\n")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les donnÃ©es
df = pd.read_json('data.json')



# Conversion de la colonne Date en format datetime
df['Date'] = pd.to_datetime(df['Date'])

# Nettoyage des valeurs non numÃ©riques dans Temperature
df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')

# Suppression des lignes avec des valeurs NaN aprÃ¨s conversion
df = df.dropna(subset=['Temperature'])



# Style des graphiques
sns.set_theme(style="darkgrid")

# CrÃ©ation des figures
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Distribution des tempÃ©ratures
sns.histplot(df["Temperature"], bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Distribution des tempÃ©ratures")

# 2. Ã‰volution temporelle de la tempÃ©rature
sns.lineplot(data=df, x="Date", y="Temperature", ax=axes[0, 1])
axes[0, 1].set_title("Ã‰volution temporelle de la tempÃ©rature")
axes[0, 1].tick_params(axis='x', rotation=45)

# 3. Matrice de corrÃ©lation
corr_matrix = df.select_dtypes(include=['float64']).corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", ax=axes[1, 0])
axes[1, 0].set_title("Matrice de corrÃ©lation")

# 4. Distribution des niveaux de COâ‚‚
sns.histplot(df["CO2_Ambient_Level"], bins=30, kde=True, ax=axes[1, 1])
axes[1, 1].set_title("Distribution des niveaux de CO2")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Charger les donnÃ©es
df = pd.read_json('../Climate_Environmental_Data.json')

# 1. DÃ©tection des doublons
doublons = df[df.duplicated()]
print(f"Doublons dÃ©tectÃ©s : \n{doublons}\n")

# 2. VÃ©rification des valeurs hors plage
df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')
df['Humidity'] = pd.to_numeric(df['Humidity'], errors='coerce')
df['Wind_Speed'] = pd.to_numeric(df['Wind_Speed'], errors='coerce')
df['CO2_Ambient_Level'] = pd.to_numeric(df['CO2_Ambient_Level'], errors='coerce')

out_of_range = df[
    (df['Temperature'] < -50) | (df['Temperature'] > 50) |
    (df['Humidity'] < 0) | (df['Humidity'] > 100) |
    (df['Wind_Speed'] < 0) | (df['Wind_Speed'] > 200) |
    (df['CO2_Ambient_Level'] < 0) | (df['CO2_Ambient_Level'] > 10000)
]
print(f"Valeurs hors plage dÃ©tectÃ©es : \n{out_of_range}\n")

# Comptage des valeurs hors plage par colonne
out_of_range_count = {
    'Temperature': ((df['Temperature'] < -50) | (df['Temperature'] > 50)).sum(),
    'Humidity': ((df['Humidity'] < 0) | (df['Humidity'] > 100)).sum(),
    'Wind_Speed': ((df['Wind_Speed'] < 0) | (df['Wind_Speed'] > 200)).sum(),
    'CO2_Ambient_Level': ((df['CO2_Ambient_Level'] < 0) | (df['CO2_Ambient_Level'] > 10000)).sum()
}

print(f"Nombre de valeurs hors plage par colonne : \n{out_of_range_count}\n")

# 3. VÃ©rification des types de donnÃ©es
print(f"Types de donnÃ©es : \n{df.dtypes}\n")

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Location'] = df['Location'].astype('category')

# 4. VÃ©rification des valeurs manquantes
missing_values = df.isnull().sum()
print(f"Valeurs manquantes par colonne : \n{missing_values}\n")

# 5. VÃ©rification des valeurs mal formatÃ©es dans la colonne 'Temperature'
non_numeric_temp = df[~df['Temperature'].apply(pd.to_numeric, errors='coerce').notna()]
print(f"Valeurs mal formatÃ©es dans la colonne 'Temperature' : \n{non_numeric_temp}\n")

num_non_numeric_temp = non_numeric_temp.shape[0]
print(f"Nombre de valeurs mal formatÃ©es dans 'Temperature' : {num_non_numeric_temp}\n")

df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')

# 6. VÃ©rification des valeurs aberrantes (outliers)
outliers_count = {}
for col in ['Temperature', 'Humidity', 'Wind_Speed', 'CO2_Ambient_Level']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outliers_count[col] = outliers.shape[0]

print(f"Nombre de valeurs aberrantes (outliers) par colonne : \n{outliers_count}\n")


```

```python
import pandas as pd

# Charger les donnÃ©es
df = pd.read_json('../data.json')

# Initialiser un dictionnaire pour stocker les valeurs mal formatÃ©es
improper_formatting = {}

# 1. VÃ©rification des colonnes numÃ©riques
for col in ['Temperature', 'Humidity', 'Wind_Speed', 'CO2_Ambient_Level']:
    # Conversion en float avec erreurs converties en NaN
    df[col + '_converted'] = pd.to_numeric(df[col], errors='coerce')
    # Filtrer les valeurs mal formatÃ©es (NaN aprÃ¨s conversion mais non NaN initialement)
    improper_formatting[col] = df[df[col + '_converted'].isna() & df[col].notna()][col]

# 2. VÃ©rification de la colonne Date
df['Date_converted'] = pd.to_datetime(df['Date'], errors='coerce')
improper_formatting['Date'] = df[df['Date_converted'].isna() & df['Date'].notna()]['Date']

# 3. VÃ©rification des colonnes catÃ©goriques (valeurs non string ou nulles)
improper_formatting['Location'] = df[~df['Location'].apply(lambda x: isinstance(x, str) or pd.isna(x))]['Location']

# 4. Affichage des rÃ©sultats
for col, values in improper_formatting.items():
    print(f"\nValeurs mal formatÃ©es dans la colonne '{col}' :\n{values if not values.empty else 'Aucune valeur mal formatÃ©e trouvÃ©e'}")
```

```python
import pandas as pd

# Charger les donnÃ©es
df = pd.read_csv('../Materials.txt')

# Fonction pour vÃ©rifier les valeurs mal formatÃ©es dans chaque colonne
def check_improper_formatting(df):
    improper_formatting_counts = {}

    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            # VÃ©rifier les valeurs non numÃ©riques dans une colonne supposÃ©e Ãªtre numÃ©rique
            non_numeric_values = df[~df[column].apply(pd.to_numeric, errors='coerce').notna()]
            improper_formatting_counts[column] = non_numeric_values.shape[0]
        else:
            # VÃ©rifier les espaces en dÃ©but ou fin pour les colonnes de type chaÃ®ne de caractÃ¨res
            improper_formatting_counts[column] = df[column].apply(lambda x: isinstance(x, str) and (x.strip() != x) if pd.notnull(x) else False).sum()

    # Afficher les rÃ©sultats
    print("\n--- Nombre de valeurs mal formatÃ©es par colonne ---")
    for column, count in improper_formatting_counts.items():
        print(f"{column}: {count} valeur(s) mal formatÃ©e(s)")

# Appeler la fonction pour gÃ©nÃ©rer le rapport sur les valeurs mal formatÃ©es
check_improper_formatting(df)
```
