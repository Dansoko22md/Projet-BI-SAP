```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
country_counts = data['Countries'].value_counts()

# Création du graphique
plt.figure(figsize=(10, 6))
sns.barplot(x=country_counts.index, y=country_counts.values, palette="viridis")
plt.xlabel("Pays")
plt.ylabel("Nombre d'entreprises")
plt.title("Répartition des entreprises par pays")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', colors=sns.color_palette("viridis"))
plt.title("Proportion of Companies by Country")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(y=country_counts.index, x=country_counts.values, palette="magma")
plt.xlabel("Number of Companies")
plt.ylabel("Country")
plt.title("Number of Companies by Country")
plt.show()
```
